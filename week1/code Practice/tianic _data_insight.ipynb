Titanic Data Insight

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

 Insights of the dataset

```python
df=sns.load_dataset('titanic')
```

```python
survival_rate=df['survived'].value_counts(normalize=True)*100
survival_rate
```

```python
sns.barplot(x='sex', y='survived', hue='pclass', data=df)
```

```python
sns.countplot(data=df, x='survived', hue='sex')
```

```python
sns.histplot(df[df['survived']==1]['age'], bins=30, kde=True, color='blue', label='Survived')
sns.histplot(df[df['survived']==0]['age'], bins=30, kde=True, color='yellow', label='Not-Survived')
plt.legend()
plt.show()
```

```python
sns.barplot(x='embarked', y='survived', hue='pclass', data=df)
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='pclass', palette=['red', 'orange', 'black'])
```

```python
sns.pairplot(df)
```

```python
df.hist(bins=10, figsize=(9,7), grid=False, color='yellow')
```
