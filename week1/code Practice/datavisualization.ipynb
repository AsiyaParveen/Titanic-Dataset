Data visualize with the help of Plot by using seaborn and MatplotLib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
df
```

```python
sns.countplot(x='survived',data=df,hue='who')
```

```python
sns.scatterplot(data=df,x='survived',y='age',hue='who')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='who')
```

```python
sns.scatterplot(data=df,x='age',y='fare',hue='who')
```

```python
plt.scatter(data=df,x="age",y='fare',c='survived') 
```

```python
sns.pairplot(df)
```

```python
sns.boxplot(data=df, x="embark_town", y="deck",hue="who")
```

```python
sns.boxenplot(data=df, x="who", y="deck",hue="embark_town")
```

```python
g = sns.FacetGrid(data=df , hue="survived" , col="pclass" , margin_titles=True ,
palette={ 2: "red", 1 :"seagreen" , 0: "gray" })
g = g.map(plt.scatter , "fare" , "age" , edgecolor="w" ).add_legend();
```

```python
sns.stripplot(data= df , x = "sex" , y = "fare" , hue= "pclass" , palette={1 :
"green" , 2 : "blue" , 3: "black"})
```

```python
corr = df.select_dtypes(include=["number"]).corr()
plt.figure(figsize=(10 , 10))
sns.heatmap(corr)
plt.title("Heat plot of Titanic data")
```

```python
df.hist(bins=5, figsize=(10,8), grid=False)
```

```python
sns.violinplot(data=df , x = "sex" , y = "fare" , hue = "class")
```
